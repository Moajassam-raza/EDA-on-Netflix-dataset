# Data cleaning

```python
df["country"] = df["country"].fillna(df["country"].mode()[0])
```

```python
df.isnull().sum()
```

```python
df["rating"] = df["rating"].fillna("not Rated")
```

```python
df.isnull().sum()
```

```python
df["duration"] = df["duration"].fillna("Unknown")
```

```python
df.isnull().sum()
```

# Removing Duplicates

```python
df.drop_duplicates(inplace=True)
```

# Exploratory Data Analysis with Visualization

**Movies vs TV Shows Count**

```python
sns.countplot(data=df, x='type')
plt.title("Movies vs TV Shows")
```

**Top Countries Producing Netflix Content**

```python
country_count = df['country'].value_counts().head(10)
sns.barplot(x=country_count.values, y=country_count.index)
plt.title("Top 10 Content-Producing Countries")
```

**Most Popular Genres**

```python
df['listed_in'].str.get_dummies(sep=', ').sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top Genres on Netflix")
```

**Ratings Distribution**

```python
sns.countplot(data=df, y='rating', order=df['rating'].value_counts().index)
plt.title("Distribution of Movie/Show Ratings")
```

**Rating distribution for Movies vs TV Shows**

```python
sns.countplot(data=df,y='rating', hue='type',
              order=df['rating'].value_counts().index)
plt.title("Ratings by Type (Movies vs TV Shows)")
plt.show()
```
